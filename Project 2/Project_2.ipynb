Lets start by looking at the neutral Supers

```python
df[df['Alignment']=='neutral']
```

Now lets look at Averages for each stat for each alignment

```python
print(f'{"Averages By Alignment":=^30}')
print(f'{"good":=^30}')
print(df[df['Alignment']=='good'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].mean())
print(f'{"neutral":=^30}')
print(df[df['Alignment']=='neutral'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].mean())
print(f'{"bad":=^30}')
print(df[df['Alignment']=='bad'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].mean())
```

Neutral has the highest stats across the board Lets dig a little deeper into why

```python
plt.figure(figsize=(10,5))
sns.ecdfplot(
    data=df,
    x='Intelligence',
    hue='Alignment'
)
plt.show()
```

It would seem the reason for neutrals having a higher average intelligence is because they have less low values 

There is a good chance that that is because of the smaller sample size

Lets Look again using a similar sample size from the other two

```python
print(f'{"Averages By Alignment With Sample Size 10"}:')
print(f'{"good":=^30}')
print(df[df['Alignment']=='good'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].sample(10).mean())
print(f'{"neutral":=^30}')
print(df[df['Alignment']=='neutral'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].sample(10).mean())
print(f'{"bad":=^30}')
print(df[df['Alignment']=='bad'][['Intelligence','Strength','Speed','Durability','Power','Combat','Total']].sample(10).mean())
```

A sample size of 10 is to small to extract meaningful information since the numbers change drastically every time the command is ran

So I will exclude the neutrals Since they have such a small sample size.
