1) Define a function that returns a Numpy Array of the Fibonacci sequence up to a given number

2) Define a function thar retruns a Numpy array of a Factorial sequence up to a given number

Bonus) Make it more efficient

```python
def fib(x):
  fibl = [0,1]

  while fibl[-1] <= x:
    fibl.append(fibl[-1] + fibl[-2])
  return np.array(fibl[:-1])

fib(0)
```

```python
def fact(x):
  factl = [1]
  y = 2
  while factl[-1] <= x:
    factl.append(factl[-1] * y)
    y = y + 1
  return np.array(factl[:-1])

fact(10)
```

```python
def fac_array(x:int) -> np.array:
    # factorial of 0 = 1
    if x == 0:
        return np.array([1])
    fac_list = [1]
    # iterate through each number up to given number adding to list
    for i in range(1,x+1):
        product = 1
        for j in range(i,0,-1):
            product *=j
        fac_list.append(product)
    # have to use array of python objects because int can't handle the size
    return np.array(fac_list)
fac_array(10)
```

```python
amz["year_added"] = amz["date_added"].str.split(", ",expand=True)[1].astype(int,errors="ignore")
```

```python
m = amz[['year_added', 'release_year']].ne(None).all(axis=1)
amz["release_gap"] = amz.loc[m, ['year_added', 'release_year']] \
    .astype(float).agg('diff', axis=1)['release_year'] \
        .reindex(amz.index, fill_value=None)
```
