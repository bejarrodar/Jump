```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
db = "sakila"
# url = "sqlite+pysqlite:///" + db
url = "mysql+mysqlconnector://root:rootroot@localhost:3306/" + db
engine = create_engine(url)
```

```python
film = pd.read_sql("SELECT * FROM film", engine, index_col = "film_id")
# film.set_index("film_id", inplace=True)
film.head()
```

```python
categories = pd.read_sql("SELECT * FROM category", engine, index_col = "category_id")
film_category = pd.read_sql("SELECT * from film_category", engine, index_col = ["category_id", "film_id"])
```

```python
film_category_id = pd.merge(left=film.reset_index(), right=film_category.reset_index(), on="film_id")
film_category_name = pd.merge(right=categories.reset_index(), left=film_category_id.reset_index(), on="category_id")
film_category_merge = film_category_name[["title", "rating", "name"]].rename(columns={"name": "category"})
film_category_merge[film_category_merge['rating'] != "PG"]
```

```python
actor = pd.read_sql("SELECT * FROM actor", engine)
film_actor = pd.read_sql("SELECT * FROM film_actor", engine)
```

```python
film_count = film_actor[["actor_id", "film_id"]].groupby("actor_id").count().rename(columns={"film_id":"film_count"})
# film_count.join(actor).sort_values("film_count", ascending=False)
pd.merge(left=film_count, right=actor, on="actor_id").sort_values("film_count", ascending=False)
```
